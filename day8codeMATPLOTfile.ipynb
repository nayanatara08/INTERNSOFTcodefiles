```python
#PLOTTING IN PANDAS
pandas = pd.Series(np.random.normal(size=10))
print(pandas)
pandas.plot()
```

```python
#Adding grids
pandas.cumsum().plot(grid=True)
```

```python
#using dataframes
df = pd.DataFrame({'a': np.random.normal(size=6), 
                       'b': np.random.normal(1, size=6), 
                       'c': np.random.normal(size=6)})
print(df)

df.cumsum(0).plot()
```

```python
#SUBPLOts
df.cumsum(0).plot(subplots=True)
```

```python
#for secondary y-axis
df.cumsum(0).plot(secondary_y='a')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 10))
for i,var in enumerate(['a','b','c']):
    df[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('y-axis')
```

```python
#saving the figuures
fig.savefig('fig.png')
```

```python
#Simple Line Plots
abc = plt.figure()
xyz = plt.axes()

nayana = np.linspace(0, 10, 25)
print(nayana)
xyz.plot(nayana, np.sin(nayana));
```

```python
#COLORS
plt.plot(nayana, np.sin(nayana - 0), color='blue')        
plt.plot(nayana, np.sin(nayana - 1), color='red')           
plt.plot(nayana, np.sin(nayana - 2), color='0.1')       
```

```python
#STYLE
plt.plot(nayana, np.sin(nayana - 0), linestyle='-')        
plt.plot(nayana, np.sin(nayana - 1), linestyle=':')           
plt.plot(nayana, np.sin(nayana - 2), linestyle='--')  
```

```python
#Axes Limits
plt.plot(nayana, np.sin(nayana - 0))

plt.xlim(0, 10)
plt.ylim(-2, 6);
```

```python
#LABELING
plt.plot(nayana, np.sin(nayana))
plt.title("My Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
kkk = np.random.RandomState(10)
x = kkk.randn(50)
y = kkk.randn(50)
colors = kkk.rand(50)
sizes = 1000 * kkk.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='Reds')
plt.colorbar() 
```

```python
#Error bars
a = np.linspace(0, 10, 30)
mm = 0.1
b = np.sin(a) + mm * np.random.randn(30)

plt.errorbar(a, b, yerr=mm, fmt='.b');
```

```python
#2-d histograms
a = [10, 4]
b = [[6, 2], [4, 2.5]]
x, y = np.random.multivariate_normal(a, b, 10)
```

```python
plt.hist2d(x, y, bins=20, cmap='rainbow')
cb = plt.colorbar()
```
